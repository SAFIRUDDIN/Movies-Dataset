```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
cols_movies=['movieid','title','genres']
cols_users=['userid','gender','age','occupation','zip']
cols_ratings=['userid','movieid','rating','timestamp']
```

```python
#Importing our data
df_movies=pd.read_csv('movies.dat',sep='::',engine='python',header=None,names=cols_movies)
df_users=pd.read_csv('users.dat',sep='::',engine='python',header=None,names=cols_users)
df_ratings=pd.read_csv('ratings.dat',sep='::',engine='python',header=None,names=cols_ratings)
```

```python
#Checking the structure of the data!
print('Numbers of rows in movies_df: ',df_movies.shape)
print('Numbers of rows in users_df: ',df_users.shape)
print('Numbers of rows in ratings_df: ',df_ratings.shape)
```

```python
# Creating a master datset : Master_Data
master_data1= df_movies.merge(df_ratings,on='movieid')
master_data=master_data1.merge(df_users,on='userid')
master_data.head()
```

```python
#Users Age Distribution
plt.figure(figsize=(6,6))
master_data.age.plot.hist()
plt.title('Users Age Distribution')
plt.xlabel('Age Factors')
plt.ylabel('Count')
plt.show()
```

```python
#Getting all unique genres
master_data.genres.unique()
```

```python
#Getting the dummies for genres
genre_dummies=master_data['genres'].str.get_dummies().add_prefix('_genre')
genre_dummies
```

```python
#Selecting the features for our model
selected_cols=['rating','gender','age']
```

```python
master_data1=master_data[selected_cols]
```

```python
#Adding the dummies to the data
master_data2=pd.concat([master_data1,genre_dummies],axis=1)
master_data2.head()
```

```python
#Getting the dummies for gender!
master_data2=pd.get_dummies(master_data2,columns=['gender'])
master_data2.head()
```

```python
#Importing Necessary Algorithms
from sklearn.neighbors.classification import KNeighborsClassifier
from sklearn.neighbors.classification import KNeighborsClassifier
from sklearn.ensemble.forest import RandomForestClassifier
from sklearn.ensemble.forest import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import MultinomialNB  
from sklearn.tree import DecisionTreeClassifier

#import libraries for model validation
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import LeaveOneOut


#import libraries for metrics and reporting 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
#I'm taking a small chunk of complete dataset as training whole dataset takes much more training time! 
import numpy as np
master_data3=master_data2.loc[np.random.choice(master_data2.index, size=15000)]
```

```python
#Checking head of master_data3  to make sure our data is randomized
master_data3.head()
```

```python
predictors=master_data3.drop(['rating'],axis=1)
target=master_data3['rating']
x_train,x_cv,y_train,y_cv=train_test_split(predictors,target,test_size=0.3,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_cv)
acc_knn = round(accuracy_score(y_pred,y_cv) * 100, 2)
print(acc_knn)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_cv)
acc_logreg = round(accuracy_score(y_pred, y_cv) * 100, 2)
print(acc_logreg)
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_cv)
acc_randomforest = round(accuracy_score(y_pred, y_cv) * 100, 2)
print(acc_randomforest)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_cv)
acc_svc = round(accuracy_score(y_pred, y_cv) * 100, 2)
print(acc_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_cv)

acc_dtc = round(accuracy_score(y_pred, y_cv) * 100, 2)
print(acc_dtc)
```

```python
models = pd.DataFrame({
    'Method': ['KNN', 'Logistic Regression', 
              'Random Forest', 'Support Vector Machine','Decision Tree Classifier'],
    'Score': [acc_knn, acc_logreg, 
              acc_randomforest, acc_svc,acc_dtc]})
models.sort_values(by='Score', ascending=False)
```

Okay! So we have just trained few models with algorithms suitable for this kind of prediction. **Logistic Regression** and **Support Vector Machine** should be preferred since they are giving good accuracy as compared to others.
